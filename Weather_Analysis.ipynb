```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wd=pd.read_csv('/content/sample_data/Weather.csv')
```

```python
wd.head(-4)
```

This data set of 8784 columns depicts the status of weather in a particular year of 365 days in an hour gap.It contains different attributes like Temperature(in C),Dew Point Temperature(in C),percent of Relative Humidity,Wind speed(in km/h),Visibility(in Km),Pressure(in Kilo pascals) and nature of weather.


```python
wd['Date/Time'] = pd.to_datetime(wd['Date/Time'])
print(wd.info())
print(wd.describe())
```

```python

plt.figure(figsize=(5, 5))
sns.lineplot(data=wd, x='Date/Time', y='Temp_C', color='blue')
plt.title("Temperature Trends Over the Year", fontsize=16)
plt.xlabel("Date/Time")
plt.ylabel("Temperature (C)")
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This graph illustrates temperature trends over the year, showing fluctuations in C across different dates. The clear periodic pattern likely reflects seasonal variations, with peaks representing warmer periods (summer) and troughs indicating colder periods (winter). The blue line helps visualize overall trends and anomalies in temperature.

```python
plt.figure(figsize=(5, 5))
sns.histplot(data=wd, x='Dew Point Temp_C', color='blue', kde=True)
plt.title("Distribution of Dew Point Temperature", fontsize=16)
plt.xlabel("Dew Point Temperature (C)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

This histogram illustrates the distribution of dew point temperatures (C) in the dataset. The bars show how frequently specific dew point temperature ranges occur, while the smooth curve provides an estimate of the underlying probability density. The graph reveals a concentration of values around 0C to 10C, indicating these are the most common dew point ranges. Extreme values at both ends (-20C to 20C) occur less frequently, reflecting rarer climatic conditions.



```python
plt.figure(figsize=(6, 6))
sns.histplot(data=wd, x='Rel Hum_%', kde=True, color='green')
plt.title("Distribution of Relative Humidity", fontsize=16)
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

This histogram shows the distribution of relative humidity percentages across the dataset. The frequencies peak between 60% and 80%, indicating that most observations fall within this range. Lower humidity values (below 40%) and higher values (above 90%) are less common. The smooth curve overlaid on the histogram highlights the general trend and density of humidity values, helping to identify the most typical conditions.


```python
plt.figure(figsize=(5, 5))
sns.lineplot(data=wd, x='Date/Time', y='Rel Hum_%', color='green')
plt.title("Relative Humidity Trends Over Time", fontsize=16)
plt.xlabel("Date/Time")
plt.ylabel("Relative Humidity (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

A line or scatter plot showing humidity levels (%) over time. Peaks and troughs in the graph highlight periods of higher or lower humidity. The visualization reflects how moisture in the air fluctuates with changing weather patterns, helping correlate it with rainfall or temperature.


```python
plt.figure(figsize=(6, 6))
sns.histplot(data=wd, x='Wind Speed_km/h', kde=True, color='orange')
plt.title("Distribution of Wind Speed", fontsize=16)
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

The histogram visualizes the distribution of wind speed, showing frequency on the y-axis and wind speed values on the x-axis. The data appears to have a right-skewed distribution, with the majority of wind speeds concentrated at lower values and tapering off as speeds increase. An orange line overlay represents a smoothed density curve, highlighting the data's skewness. This visualization suggests that extreme wind speeds are less common compared to moderate ones.

```python
plt.figure(figsize=(6, 5))
sns.lineplot(data=wd, x='Date/Time', y='Wind Speed_km/h', color='orange')
plt.title("Wind Speed Trends Over Time", fontsize=16)
plt.xlabel("Date/Time")
plt.ylabel("Wind Speed (km/h)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This time-series plot depicts wind speed trends over a specified period, with wind speed (in km/h) on the y-axis and date/time on the x-axis. The data shows consistent fluctuations in wind speed, with occasional spikes reaching around 80 km/h. No clear seasonal pattern is visible, but the variability suggests dynamic weather conditions over the observed timeframe. Peaks may correspond to extreme weather events or high wind periods.

```python
plt.figure(figsize=(6, 6))
sns.histplot(data=wd, x='Press_kPa', kde=True, color='red')
plt.title("Distribution of Pressure", fontsize=16)
plt.xlabel("Pressure (kPa)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

This histogram illustrates the distribution of atmospheric pressure, with frequency on the y-axis and pressure values on the x-axis. The data appears to follow a bell-shaped curve, resembling a normal distribution, with most values concentrated around the central peak. The red line overlay represents a smoothed density curve, further emphasizing the symmetric nature of the distribution. This pattern suggests that pressure values are generally stable, with only a few extreme deviations.

```python
plt.figure(figsize=(6, 5))
sns.lineplot(data=wd, x='Date/Time', y='Press_kPa', color='red')
plt.title("Pressure Trends Over Time", fontsize=16)
plt.xlabel("Date/Time")
plt.ylabel("Pressure (kPa)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This line chart illustrates the trend of atmospheric pressure over time, with the x-axis showing the date and the y-axis representing pressure in kilopascals (kPa). The data exhibits frequent fluctuations within a range of ~98 to ~103 kPa. While the overall pattern appears stable, notable spikes and dips are observed, particularly a sharp rise in late 2012 and a significant drop near the end of the year. These variations reflect natural atmospheric changes, likely influenced by weather systems and seasonal transitions.








```python
plt.figure(figsize=(6, 6))
sns.lineplot(data=wd, x='Date/Time', y='Visibility_km', color='blue')
plt.title("Visibility Trends Over Time", fontsize=16)
plt.xlabel("Date and Time", fontsize=12)
plt.ylabel("Visibility (km)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

This line chart illustrates the trend of visibility over time, with the x-axis showing the date and the y-axis representing visibility in kilometers (km). The data exhibits frequent fluctuations within a range of ~0 to ~50 km. While the overall pattern appears dynamic, periods of low visibility are observed, indicating conditions like fog, heavy precipitation, or haze. The variability reflects natural weather changes, likely influenced by factors such as humidity, temperature, and seasonal transitions.

```python
numeric_wd = wd.select_dtypes(include=['number'])
print(numeric_wd.head())
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_wd.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

Strong Positive Correlation:
       Temp_C and Dew Point Temp_C show a high positive correlation(0.93), indicating that as temperature increases, dew point temperature also rises.

Moderate Negative Correlation:
      Rel Hum_% and Visibility_km are negatively correlated (0.63), suggesting that higher relative humidity often corresponds to reduced visibility.

Weak Correlation:
      Variables like Wind Speed_km/h and Visibility_km (0.005) or Rel Hum_% and Temp_C (0.22) show little to no linear relationship.

Independent Variations:
       Many values close to zero indicate that changes in certain parameters, like Wind Speed and Dew Point, do not strongly affect one another.

```python
sns.pairplot(wd)
plt.suptitle("Pair Plot of All Weather Variables", fontsize=16)
plt.tight_layout()
plt.show()
```

 The diagonal elements show a distribution plot (like histograms) which represents the distribution of each individual variable in the dataset.

 Pair plots help us to identify trends or relationships between variables, such as linear, nonlinear, or no relationships at all.

 By looking at the scatterplots, you can quickly spot if two variables have a strong or weak relationship, and the direction (positive or negative) of the relationship.
